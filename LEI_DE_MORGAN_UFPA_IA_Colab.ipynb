A Lei de Morgan e importante porque permite transformar negacoes de expressoes compostas em formas equivalentes mais simples, algo muito utilizado em matematica discreta, programacao e eletronica digital.


## Slide 2 - Revisao de logica booleana

Na algebra booleana, trabalhamos com duas possibilidades logicas: verdadeiro e falso.

### Operadores basicos

- `AND` (`and` em Python): so resulta em verdadeiro quando as duas proposicoes sao verdadeiras
- `OR` (`or` em Python): resulta em verdadeiro quando pelo menos uma proposicao e verdadeira
- `NOT` (`not` em Python): inverte o valor logico da proposicao

### Escrita matematica

- $A \land B$ representa a conjuncao logica
- $A \lor B$ representa a disjuncao logica
- $\neg A$ representa a negacao de $A$

Esses operadores servem de base para a formulacao e a demonstracao das Leis de De Morgan.


## Slide 3 - Primeira Lei de De Morgan

A primeira lei afirma que a negacao de uma conjuncao e equivalente a disjuncao das negacoes:

$$\neg (A \land B) = (\neg A) \lor (\neg B)$$

### Interpretacao conceitual

Negar a afirmacao `A e B` significa dizer que pelo menos uma das duas proposicoes falhou. Em outras palavras, ou `A` e falsa, ou `B` e falsa, ou ambas sao falsas.

Essa equivalencia e muito usada para reescrever condicoes logicas em programas e para substituir combinacoes de portas em circuitos.


## Slide 4 - Segunda Lei de De Morgan

A segunda lei afirma que a negacao de uma disjuncao e equivalente a conjuncao das negacoes:

$$\neg (A \lor B) = (\neg A) \land (\neg B)$$

### Interpretacao conceitual

Negar a afirmacao `A ou B` significa exigir que nenhuma das duas proposicoes seja verdadeira. Portanto, para que a expressao negada seja verdadeira, `A` precisa ser falsa e `B` tambem precisa ser falsa.

Essa forma tambem e essencial em simplificacao de expressoes booleanas e na analise de tabelas-verdade.


## Slide 5 - Tabela-verdade da primeira lei

### Instrucoes para executar

- esta celula gera a tabela-verdade da expressao $\neg(A \land B)$ e a compara com $(\neg A) \lor (\neg B)$
- execute a celula de codigo logo abaixo para verificar todas as combinacoes possiveis de `A` e `B`
- observe que a coluna da esquerda e a coluna da direita coincidem em todas as linhas
- se quiser adaptar o exemplo, voce pode alterar a funcao de exibicao de `V` e `F` ou mudar a mensagem final


```python
from itertools import product

def vf(valor: bool) -> str:
    return 'V' if valor else 'F'

print('Slide 5 - Verificacao da primeira equivalencia da Lei de Morgan')
print('-' * 66)
print(f"{'A':^5}{'B':^5}{'not (A and B)':^18}{'(not A) or (not B)':^22}{'Iguais?':^10}")
print('-' * 66)

equivalente = True
for A, B in product([False, True], repeat=2):
    esquerda = not (A and B)
    direita = (not A) or (not B)
    iguais = esquerda == direita
    equivalente = equivalente and iguais
    print(f"{vf(A):^5}{vf(B):^5}{vf(esquerda):^18}{vf(direita):^22}{vf(iguais):^10}")

print('-' * 66)
print('Conclusao:', 'a equivalencia foi confirmada.' if equivalente else 'ha divergencia na comparacao.')
```

## Slide 6 - Tabela-verdade da segunda lei

### Instrucoes para executar

- esta celula compara a expressao $\neg(A \lor B)$ com $(\neg A) \land (\neg B)$
- execute a celula logo abaixo para montar a tabela-verdade completa
- confira se as duas colunas centrais produzem exatamente os mesmos resultados
- para personalizar a experiencia, voce pode alterar os rotulos usados na exibicao ou expandir a tabela com comentarios adicionais


```python
from itertools import product

def vf(valor: bool) -> str:
    return 'V' if valor else 'F'

print('Slide 6 - Verificacao da segunda equivalencia da Lei de Morgan')
print('-' * 66)
print(f"{'A':^5}{'B':^5}{'not (A or B)':^18}{'(not A) and (not B)':^22}{'Iguais?':^10}")
print('-' * 66)

equivalente = True
for A, B in product([False, True], repeat=2):
    esquerda = not (A or B)
    direita = (not A) and (not B)
    iguais = esquerda == direita
    equivalente = equivalente and iguais
    print(f"{vf(A):^5}{vf(B):^5}{vf(esquerda):^18}{vf(direita):^22}{vf(iguais):^10}")

print('-' * 66)
print('Conclusao:', 'a equivalencia foi confirmada.' if equivalente else 'ha divergencia na comparacao.')
```

## Slide 7 - Portas logicas e equivalencia de circuitos

Na eletronica digital, as Leis de De Morgan mostram que certos circuitos podem ser reescritos com outro conjunto de portas sem alterar o comportamento logico.

### Leitura textual das equivalencias

- negar a saida de uma porta `AND` equivale a negar cada entrada e ligar essas entradas a uma porta `OR`
- negar a saida de uma porta `OR` equivale a negar cada entrada e ligar essas entradas a uma porta `AND`

### Consequencia pratica

Essas transformacoes ajudam no projeto de circuitos, principalmente quando se deseja implementar uma funcao apenas com familias especificas de portas, como `NAND` ou `NOR`.

Assim, a Lei de De Morgan nao e apenas uma identidade simbolica: ela tambem representa equivalencia funcional entre circuitos digitais.


## Slide 8 - Simplificacao de expressoes booleanas

### Instrucoes para executar

- a celula abaixo testa algumas expressoes equivalentes obtidas com a Lei de De Morgan
- execute a celula para verificar, em todas as combinacoes de `A` e `B`, que as formas original e simplificada produzem o mesmo resultado
- observe como a regra ajuda a reescrever expressoes mais longas em formas mais faceis de interpretar
- se quiser experimentar, adicione novos pares de expressoes seguindo o mesmo padrao do dicionario `exemplos`


```python
from itertools import product

exemplos = {
    'Primeira lei': (
        lambda A, B: not (A and B),
        lambda A, B: (not A) or (not B),
        'not (A and B)',
        '(not A) or (not B)'
    ),
    'Segunda lei': (
        lambda A, B: not (A or B),
        lambda A, B: (not A) and (not B),
        'not (A or B)',
        '(not A) and (not B)'
    ),
    'Absorcao apos simplificacao': (
        lambda A, B: not ((A and B) or A),
        lambda A, B: not A,
        'not ((A and B) or A)',
        'not A'
    )
}

for titulo, (expressao_1, expressao_2, texto_1, texto_2) in exemplos.items():
    print(f'\n{titulo}')
    print(f'Comparando: {texto_1}  <=>  {texto_2}')
    equivalentes = True
    for A, B in product([False, True], repeat=2):
        resultado_1 = expressao_1(A, B)
        resultado_2 = expressao_2(A, B)
        equivalentes = equivalentes and (resultado_1 == resultado_2)
        print(f'A={A:<5} B={B:<5} original={resultado_1:<5} simplificada={resultado_2:<5}')
    print('Equivalencia confirmada.' if equivalentes else 'Equivalencia nao confirmada.')
```

## Slide 9 - Aplicacao pratica em Python

### Instrucoes para executar

- a celula abaixo simula um sistema simples de monitoramento de laboratorio
- o ambiente so e considerado `totalmente seguro` quando a porta esta fechada e a camera esta online
- a condicao de alerta e escrita primeiro como `not (porta_fechada and camera_online)` e depois reescrita com a primeira Lei de De Morgan
- execute a celula e observe como a forma equivalente ajuda a identificar de maneira direta qual condicao falhou
- para explorar novos cenarios, altere os valores da lista `cenarios`


```python
cenarios = [
    {'nome': 'Cenario 1', 'porta_fechada': True, 'camera_online': True},
    {'nome': 'Cenario 2', 'porta_fechada': False, 'camera_online': True},
    {'nome': 'Cenario 3', 'porta_fechada': True, 'camera_online': False},
    {'nome': 'Cenario 4', 'porta_fechada': False, 'camera_online': False},
]

print('Slide 9 - Sistema de monitoramento')
print('-' * 72)

for cenario in cenarios:
    porta_fechada = cenario['porta_fechada']
    camera_online = cenario['camera_online']

    alerta_forma_1 = not (porta_fechada and camera_online)
    alerta_forma_2 = (not porta_fechada) or (not camera_online)

    print(cenario['nome'])
    print(f"  porta_fechada = {porta_fechada}")
    print(f"  camera_online = {camera_online}")
    print(f"  alerta usando not (A and B): {alerta_forma_1}")
    print(f"  alerta usando (not A) or (not B): {alerta_forma_2}")

    if alerta_forma_2:
        motivos = []
        if not porta_fechada:
            motivos.append('porta aberta')
        if not camera_online:
            motivos.append('camera offline')
        print('  motivo do alerta:', ', '.join(motivos))
    else:
        print('  ambiente seguro: nenhuma falha detectada')

    print('-' * 72)
```
